```julia
α = 2
Δx = 1

Δt = (Δx^2)/(4α)
γ = (α*Δt)/(Δx^2)
```

```julia
using Plots
```

```julia
us = Array{Float64, 3}(undef, 50, 50, 750);
```

```julia
fill!(us, 0)
us[:, 1, :] .= 0
us[:, 50, :] .= 0
us[50, :, :] .= 0
us[1, :, :] .= 100;
```

```julia
us
```

```julia
function calculo(u)
    for κ in 1:749
        for i in 2:49
            for j in 2:49
                u[i, j, κ + 1] = γ*(u[i + 1, j, κ] .+ u[i - 1, j, κ] .+ u[i, j + 1, κ] .+ u[i, j - 1, κ] .- u[i, j, κ]) .+ u[i, j, κ]
            end
        end
    end
    return u
end
```

```julia
calculo(us)
```

```julia
anim = @animate for i ∈ 1:750
    heatmap(us[:, :, i])
end
gif(anim, "anim_fps15.gif", fps = 15)
```
