```julia
using PyCall
```

```julia
using PyPlot
```

```julia
R = 500
```

```julia
us = Array{Float64, 3}(undef, 2001, 2001, 3);
```

```julia
fill!(us, 1)
us[:, 1, :] .= 0
us[:, 2001, :] .= 0
us[2001, :, :] .= 0
us[1, :, :] .= 0;
```

```julia
is = (1000-R):(1000+R)
jsmas = [Int(round(1000 + sqrt(R^2 - (i-1000)^2))) for i in is]
jsmenos = [Int(round(1000 - sqrt(R^2 - (i-1000)^2))) for i in is];
```

```julia
A = minimum(jsmas)
B = maximum(jsmenos)
for i in is
    for j in jsmas
        for l in A:j
            us[i, l, 1] = 3
        end
    end
    for j in jsmenos
        for l in j:B
            us[i, l, 1] = 3
        end
    end
end
```

```julia
sns = pyimport("seaborn")
```

```julia
ax = sns.heatmap(us[:, :, 1])
```
