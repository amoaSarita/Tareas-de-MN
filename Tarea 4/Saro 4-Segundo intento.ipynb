```julia
us = Array{Float64, 3}(undef, 10, 10, 3);
```

```julia
fill!(us, 1)
us[:, 1, :] .= 0
us[:, 10, :] .= 0
us[10, :, :] .= 0
us[1, :, :] .= 100;
```

```julia
us
```

```julia
function calculo(u)
    for κ in 1:2
        for i in 2:9
            for j in 2:9
                u[i, j, κ + 1] = 1*(u[i + 1, j, κ] .+ u[i - 1, j, κ] .+ u[i, j + 1, κ] .+ u[i, j - 1, κ] .- u[i, j, κ]) .+ u[i, j, κ]
            end
        end
    end
    return u
end
```

```julia
calculo(us);
```
