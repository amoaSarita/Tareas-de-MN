```julia
using Plots
```

```julia
R = 15
```

```julia
us = Array{Float64, 3}(undef, 201, 201, 3);
```

```julia
fill!(us, 1)
us[:, 1, :] .= 0
us[:, 201, :] .= 0
us[201, :, :] .= 0
us[1, :, :] .= 100;
```

```julia
is = (100-R):(100+R)
jsmas = [Int(round(100 + sqrt(R^2 - (i-100)^2))) for i in is]
jsmenos = [Int(round(100 - sqrt(R^2 - (i-100)^2))) for i in is];
```

```julia
A = minimum(jsmas)
B = maximum(jsmenos)
for i in is
    for j in jsmas
        for l in A:j
            us[i, l, 1] = 3
        end
    end
    for j in jsmenos
        for l in j:B
            us[i, l, 1] = 3
        end
    end
end
```

```julia
heatmap(us[:, :, 1])
```
